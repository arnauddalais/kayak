```python
import pandas as pd
import json
```

```python
# read the weather csv and load booking json
df_weather = pd.read_csv('src/cities_weather.csv')
df_booking = pd.DataFrame(json.load(open('booking/booking/booking.json')))
```

```python
#checking dataframe
df_weather.head()
```

```python
#checking dataframe
df_booking.head(20)
```

```python
# create seaparete columns for latitude and longitude for hotels and translaste score as type float
df_booking['lat_hotel'] = df_booking['latlon'].apply(lambda x : x.split(',')[0])
df_booking['lon_hotel'] = df_booking['latlon'].apply(lambda x : x.split(',')[1])
df_booking.drop("latlon", axis=1, inplace=True)
df_booking['lat_hotel'] = df_booking['lat_hotel'].astype('float')
df_booking['lon_hotel'] = df_booking['lon_hotel'].astype('float')
df_booking['score'] = df_booking['score'].apply(lambda x : float(x.replace(',','.')))
df_booking.head()
```

```python
# merging the both files with the key id_city
df_kayak = df_booking.merge(df_weather, how='left', on='id_city')
```

```python
# Basic stats
print("Number of rows : {}".format(df_kayak.shape[0]))
print()

print("Display of dataset: ")
display(df_kayak.head())
print()

print("Basics statistics: ")
data_desc = df_kayak.describe(include='all')
display(data_desc)
print()

print("Percentage of missing values: ")
display(100*df_kayak.isnull().sum()/df_kayak.shape[0])
```

```python
#dropping useless columns (twice)
df_kayak.drop("city", axis=1, inplace=True)
```

```python
# save file to csv
df_kayak.to_csv('src/kayak.csv', index=False)
```
